Os seguintes campos :

Após essa verificação , será feito o drop dessas linhas com os valores distoantes, as linhas duplicadas e valores faltantes

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[df_adultData['fnlwgt'] !=' C']
df_adultData['fnlwgt'].value_counts() # valor ' ?' identificado - tem um espaço
```

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[df_adultData['native-country'] !=' ?']
df_adultData['native-country'].value_counts() # valor ' ?' identificado - tem um espaço
```

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[df_adultData['workclass'] !=' ?']
df_adultData['workclass'].value_counts() # valor ' ?' identificado - tem um espaço
```

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[df_adultData['occupation'] !=' ?']
df_adultData['occupation'].value_counts() # valor ' ?' identificado - tem um espaço
```

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[df_adultData['capital-gain'] !=' A']
df_adultData['capital-gain'].value_counts(ascending=True) # valor A identificado
```

```python
#Como são poucos registros distoantes, droparemos as linhas com esses valores

df_adultData = df_adultData[(df_adultData['age'] !='B') | (df_adultData['capital-gain'] !='D')]
df_adultData['age'].value_counts(ascending=True) # valor A identificado
```

```python
# Dropando Linhas Duplicadas

df_adultData=df_adultData.drop_duplicates()
```

```python
# removendo valores vazios e NaN

df_adultData.dropna(inplace=True)
```

```python
df_adultData.shape
```

Alterando dtypes

```python
# df_adultData['fnlwgt'].astype('int64')
# aqui acabei verificando a existencia de mais um valor que não tinha reparado anteriormente
```

```python
df_adultData= df_adultData.astype({
    'age':int,
    'fnlwgt':int,
    'education-num':int,
    'hours-per-week':int})
```

```python
df_adultData=df_adultData.astype({
    'capital-gain':float,
    'capital-loss':float
})
```

```python
#verificando novos tipos de dados

df_adultData.dtypes
```

```python
# Resetando indexes depois das alterações
df_adultData= df_adultData.reset_index(drop=True)
```
